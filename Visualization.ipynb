```python
# Anil Kumar Ravuru

import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

source, dest, weights = [], [], []
with open('sample_graph.txt', 'r') as lines:
    src = 0
    for line in lines:
        w = list(map(int, line.split(' ')))
        for i in range(src+1, len(w)):
            if w[i] != 0:
                source.append(src)
                dest.append(i)
                weights.append(w[i])
        src += 1

groups = [0]*src
with open('sample_graph_groups.txt', 'r') as lines:
    grp_num = 0
    for line in lines:
        nodes = list(map(int, line.split(' ')))
        for num in nodes:
            groups[num] = grp_num
        grp_num += 1
        
plt.figure(figsize=(50,30))
graph_df = pd.DataFrame({ 'source': source, 'dest': dest, 'weights': weights})
G = nx.from_pandas_edgelist(graph_df, 'source', 'dest', ['weights'], create_using=nx.Graph() )
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color=groups, cmap=plt.cm.Set1, node_size=5000, font_size=50)
edge_lab = {}
for i in range(len(source)):
    edge_lab[(source[i], dest[i])] = str(weights[i])
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_lab, font_color='red', font_size=30)
plt.axis('off')
plt.show()
```
